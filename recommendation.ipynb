```python
user_data=data[(data.user==23768)]
user_data
```

```python
user_rating=user_data[['rating']]
user_rating
```

```python
features=user_data.drop(['rating','author','Publisher','user','ISBN','year'], axis = 1)
```

```python
print('Shape of rating is:',user_rating.shape)
print('Shape of features is:',features.shape)
```

```python
user_profile = features.T.dot(user_rating.rating)
user_profile
```

```python
data = data.set_index(data.ISBN)

# let's view the head
data.head()
```

```python
data.drop(['user','ISBN','year','rating','author','Publisher'], axis=1, inplace=True)

data.head()
```

```python
recommendation_table_df = (data.dot(user_profile)) / user_profile.sum()

# Let's view the recommendation table
recommendation_table_df.head()
```

```python
recommendation_table_df.sort_values(ascending=False, inplace=True)
#recommendation_table_df.drop_duplicates(keep ='last', inplace = True)
#Just a peek at the values
recommendation_table_df
```

```python
index = recommendation_table_df.index
is_duplicate = index.duplicated(keep="first")
not_duplicate = ~is_duplicate
no_duplicate_indeces = recommendation_table_df[not_duplicate]
no_duplicate_indeces
```
